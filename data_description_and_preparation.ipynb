# Introduction of data

The data contains 163212 rows (observations) and 18 columns (variables). Specifically,



*   id: Policy identification
*   expo: exposure, defined in yearly unit, within (0, 1]

* claim: binary indicator, take only value 0 (the policy has no claim) or 1 (the policy has at least 1 claim)
*   nclaims: actual number of claims from a policy during the exposure


*   amounts: total claim severity of all claims occured during the exposure period
*   average: average amount of claim severity


*  coverage:factor data type with 3 levels: TPL, TPL+, TPL++.
*   ageph: age of the policy in years


*  sex: factor data type with 2 levels: male and female.
*  bm: bonus-malus level


*   agec: age of the car in years
*   fuel: factor data type with 2 levels: "gasoline","diesel".

*   use: factor data type with 2 levels: "private", "work"
*   lat, long: Geographic location of the insured vehicle


*   power: engine power of the car
*   List item

*   fleet:factor data type, indicates whether the vehicle belongs to a fleet policy, with 2 levels: 1 (yes) and 0 (no)
*   postcode: Residential area of the policy
